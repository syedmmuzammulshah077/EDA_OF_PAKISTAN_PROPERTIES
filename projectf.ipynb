## Before we do EDA, lets separate Numerical and categorical variables for easy analysis

```python
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

## Step 9: EDA Univariate Analysis

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'property_type', data = data, color = 'blue', 
              order = data['property_type'].value_counts().index)
sns.countplot(ax = axes[0, 1], x = 'city', data = data, color = 'blue', 
              order = data['city'].value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'province_name', data = data, color = 'blue', 
              order = data['province_name'].value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'purpose', data = data, color = 'blue', 
              order = data['purpose'].value_counts().index);
sns.countplot(ax = axes[2, 0], x = 'price', data = data, color = 'blue', 
              order = data['price'].value_counts().index);
sns.countplot(ax = axes[2, 1], x =  'Area Size', data = data, color = 'blue', 
              order = data ['Area Size'].value_counts().index);
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```

```python
fig, axarr = plt.subplots(4, 2, figsize=(12, 18))
data.groupby('location')['price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("location Vs price", fontsize=18)
data.groupby('city')['price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title("city Vs price", fontsize=18)
data.groupby('province_name')['price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("province_name Vs price", fontsize=18)
data.groupby('Area Type')['price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("Area Type Vs price", fontsize=18)
data.groupby('purpose')['price'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title("purpose Vs price", fontsize=18)
data.groupby('Area Size')['price'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title("Area Size Vs price", fontsize=18)
data.groupby('baths')['price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title("baths Vs price", fontsize=18)
data.groupby('bedrooms')['price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12)
axarr[3][1].set_title("bedrooms Vs price", fontsize=18)
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```

## Step 10: EDA Multivariate Analysis

```python
numerical_columns = ['price', 'latitude', 'longitude', 'baths', 'bedrooms','Area Size']

# Calculate the correlation matrix
correlation_matrix = data[numerical_columns].corr()

```

```python
# Plotting the heatmap
plt.figure(figsize=(12, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
